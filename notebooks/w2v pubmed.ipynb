```python
lines_test=[]
for line in open('pubmed (1).txt', 'r'):
    lines_test.append(line.strip())
```

```python
from gensim.scripts.glove2word2vec import glove2word2vec
glove_input_file = 'glove.6B/glove.6B.100d.txt'
word2vec_output_file = 'word2vec.txt'
glove2word2vec(glove_input_file, word2vec_output_file)
```

```python
from gensim.models import KeyedVectors
# load the Stanford GloVe model
filename = 'word2vec.txt'
model_glove = KeyedVectors.load_word2vec_format(filename, binary=False)
```

```python
from gensim.test.utils import common_texts, get_tmpfile
from gensim.models import Word2Vec
```

```python
import nltk
from nltk.corpus import stopwords
stopw=set(stopwords.words('english'))
```

```python
sentences = [line.lower().strip().replace('[', '').replace(']', '').replace('(', '')\
             .replace(')', '').replace('.', '').split(' ') for line in lines_test]

sentences= [[w for w in s if w not in stopw] for s in sentences ]
# train model
model = Word2Vec(sentences, size=100, window=5, min_count=1, workers=4)
# summarize the loaded model
print(model)
# summarize vocabulary
words = list(model.wv.vocab)
print(words)
# access vector for one word
print(model['sentence'])
# # save model
# model.save('model.bin')
# # load model
# new_model = Word2Vec.load('model.bin')
# print(new_model)
```

```python
import numpy as np
```

```python
d_ce=[]
for i in pubmed_ce:
    t_c=i[0].lower().strip().replace('[', '').replace(']', '').replace('(', '').replace(')', '').replace('.', '').split(' ')
    t_e=i[1].lower().strip().replace('[', '').replace(']', '').replace('(', '').replace(')', '').replace('.', '').split(' ')
    tcl=[]
    for word in t_c:
        try:
            tcl.append(model.wv[word])
        except:
            tcl.append(model_glove.wv[word])
    tel=[]
    for word_e in t_e:
        try:
            tel.append(model.wv[word_e])
        except:
            tel.append(model_glove.wv[word_e])
        
    tcl_s=sum(tcl)/len(tcl)
    tel_s=sum(tel)/len(tel)
    d_ce.append(np.linalg.norm(tel_s-tcl_s))
```

```python
sum(d_ce)/len(d_ce)
```
